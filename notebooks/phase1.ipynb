```python
import json

file_path = "../data/raw/sample_candidates.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

print("Number of candidates:", len(data))
```

```python
candidate = data[0]
candidate
```

```python
# What keys does every candidate have?

for key in data[0]:
    print(key)
```

```python
# Check whether every candidate has the same top-level keys

all_same = all(set(candidate.keys()) == set(data[0].keys()) for candidate in data)

print("All candidates have same structure:", all_same)
```

```python
candidate = data[0]

for key, value in candidate.items():
    print(f"{key}: {type(value)}")
```

```python
candidate = data[0]

for key, value in candidate.items():
    print(f"{key}")
    print(f"Type : {type(value)}")

    if isinstance(value, list):
        print(f"Length : {len(value)}")

    print("-" * 40)
```

```python
profile = candidate["profile"]

for key, value in profile.items():
    print(f"{key} : {value}")
```

```python
career = candidate["career_history"]

print("Number of jobs:", len(career))

for i, job in enumerate(career):
    print("\n" + "="*60)
    print(f"Job {i+1}")
    print("="*60)

    for key, value in job.items():
        print(f"{key}: {value}")
```

```python
skills = candidate["skills"]

print("Total skills:", len(skills))

print("\nFirst skill:\n")
print(skills[0])
```

```python
for skill in skills:
    print(skill)
```

```python
signals = candidate["redrob_signals"]

for key, value in signals.items():
    print(f"{key}: {value}")
```
